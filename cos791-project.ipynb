```python
import cv2
import os
import matplotlib.pyplot as plt

# Path to the original video
input_video_path = "/kaggle/input/cos791-video-frames-labels/videos/Hockey1.mp4-.mp4"  # Original video path
output_dir = "runs/detect/predict3/"  # Directory where output frames are saved
output_video_path = "/kaggle/working/Hockey1_output_video.mp4"  # Path to save the final video

# Read the original video to extract FPS and frame size
cap = cv2.VideoCapture(input_video_path)

if not cap.isOpened():
    print("Error: Could not open the video file.")
    exit()

# Get the original video's FPS, frame width, and frame height
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Initialize the VideoWriter object with the same FPS and frame size as the original video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for mp4
out = cv2.VideoWriter(output_video_path, fourcc, fps, (640, 640))  # Ensure consistent frame size

# Iterate over the saved output frames and write them into the video
frame_files = sorted(os.listdir(output_dir), key=lambda x: int(x.split('.')[0].split('_')[1]))  # Ensure sorted order

# Function to display frames using matplotlib (for environments without OpenCV GUI support)
def display_frame_with_matplotlib(frame):
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off the axis
    plt.show()

for frame_file in frame_files:
    frame_path = os.path.join(output_dir, frame_file)
    
    # Read each frame
    frame = cv2.imread(frame_path)
    
    if frame is None:
        print(f"Error reading frame {frame_path}")
        continue

    # Resize the frame to match the output video size (640x640)
    frame = cv2.resize(frame, (640, 640))
    
    # Visualize the frame using OpenCV (for local environments)
    # Uncomment below if you are running this locally with OpenCV GUI support
    # cv2.imshow('Frame', frame)
    # cv2.waitKey(1)  # Display the frame for 1 ms

    # Visualize the frame using matplotlib (for environments like Kaggle)
#     display_frame_with_matplotlib(frame)
    
    # Write the frame into the video
    out.write(frame)

# Release the video writer and original video capture
out.release()
cap.release()

# Close any OpenCV windows (if using imshow)
# Uncomment below if you are running this locally with OpenCV GUI support
# cv2.destroyAllWindows()

print(f"Video saved at: {output_video_path}")
```
