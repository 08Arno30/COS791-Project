```python
# Import required libraries
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model
# model = YOLO('yolov8s.pt')  # Use an appropriate model size based on your needs

# Or load pretrained model
model = YOLO("/kaggle/working/runs/detect/train2/weights/best.pt")  # Path to your trained model weights

# Training configuration
data_yaml_path = '/kaggle/working/icehockey_girlshockey_data.yaml'  # Path to your dataset YAML file
epochs = 150  # Set the number of epochs
img_size = 640  # Set the image size for training
batch_size = 32  # Set the batch size

# Train the model with optional WandB logging
model.train(data=data_yaml_path, epochs=epochs, imgsz=img_size, batch=batch_size)
```
