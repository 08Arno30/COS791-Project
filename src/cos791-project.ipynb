### Tracker Class
The following Tracker class was adapted from this link. The author used the class to fix a bug.

We follow the same approach for implementing the DeepSort algorithm for object tracking.

```python
!git clone https://github.com/nwojke/deep_sort.git
%cd deep_sort
```

```python
from deep_sort.deep_sort.tracker import Tracker as DeepSort
from deep_sort.tools import generate_detections as generate_encoder
from deep_sort.deep_sort import nn_matching
from deep_sort.deep_sort.detection import Detection
import numpy as np


class ObjectTracker:
    tracking_system = None
    feature_encoder = None
    current_tracks = None

    def __init__(self):
        max_distance = 0.4
        nn_budget_limit = None

        encoder_model_path = 'model_data/mars-small128.pb'

        distance_metric = nn_matching.NearestNeighborDistanceMetric("cosine", max_distance, nn_budget_limit)
        self.tracking_system = DeepSort(distance_metric)
        self.feature_encoder = generate_encoder.create_box_encoder(encoder_model_path, batch_size=1)

    def update(self, frame, detected_objects):
        if len(detected_objects) == 0:
            self.tracking_system.predict()
            self.tracking_system.update([])  
            self.refresh_tracks()
            return

        bounding_boxes = np.asarray([obj[:-1] for obj in detected_objects])
        bounding_boxes[:, 2:] = bounding_boxes[:, 2:] - bounding_boxes[:, 0:2]
        detection_scores = [obj[-1] for obj in detected_objects]

        encoded_features = self.feature_encoder(frame, bounding_boxes)

        detections = []
        for i, bbox in enumerate(bounding_boxes):
            detections.append(Detection(bbox, detection_scores[i], encoded_features[i]))

        self.tracking_system.predict()
        self.tracking_system.update(detections)
        self.refresh_tracks()

    def refresh_tracks(self):
        active_tracks = []
        for track in self.tracking_system.tracks:
            if not track.is_confirmed() or track.time_since_update > 1:
                continue
            box_coordinates = track.to_tlbr()

            track_id = track.track_id

            active_tracks.append(TrackedObject(track_id, box_coordinates))

        self.current_tracks = active_tracks


class TrackedObject:
    track_id = None
    bounding_box = None

    def __init__(self, track_id, bounding_box):
        self.track_id = track_id
        self.bounding_box = bounding_box
```

### Object tracking
The code below will make use of our fine-tuned YOLOv8 model, trained to detect very small hockey pucks or balls and track it through a video.

```python
import os
import random
import cv2

video_names = ['Hockey0.mp4', 'Hockey1.mp4-.mp4']
selected_video = video_names[0]

video_path = f'/kaggle/input/cos791-video-frames-labels/videos/{selected_video}'

# create ouput directory
if not os.path.exists('/kaggle/working/output_video'):
    os.makedirs('/kaggle/working/output_video', exist_ok=True)
    
video_out_path = os.path.join('/kaggle/working/output_video', f'{selected_video}_out.mp4')

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()

cap_out = cv2.VideoWriter(video_out_path, cv2.VideoWriter_fourcc(*'MP4V'), cap.get(cv2.CAP_PROP_FPS),
                          (frame.shape[1], frame.shape[0]))

# Load fine-tuned model
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

tracker = Tracker()

bbox_color = (0, 255, 0)

detection_threshold = 0.5

while ret:

    # retrieve model predictions
    results = model(frame)

    for result in results:
        detections = []
        for r in result.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = r
            x1 = int(x1)
            x2 = int(x2)
            y1 = int(y1)
            y2 = int(y2)
            class_id = int(class_id)
            if score > detection_threshold:
                detections.append([x1, y1, x2, y2, score])

        tracker.update(frame, detections)

        for track in tracker.tracks:
            bbox = track.bbox
            x1, y1, x2, y2 = bbox
            track_id = track.track_id

            # draw bounding box
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), bbox_color, 2)

    # write the frame to the video
    cap_out.write(frame)
    
    # read the next frame
    ret, frame = cap.read()

cap.release()
cap_out.release()
cv2.destroyAllWindows()
```
